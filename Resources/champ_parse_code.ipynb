```python
import requests
import json
import pandas as pd
```

```python
champ_url = "https://ddragon.leagueoflegends.com/cdn/11.5.1/data/en_US/champion.json"

response = requests.get(champ_url)
results = response.json()


champs = results['data']

champs

```

```python
champ_dict = {
    "Name": [],
    "Key": [],
    "Role": []
}

for x in champs:
    champ_dict['Name'].append(champs[x]['id'])
    champ_dict['Key'].append(champs[x]['key'])
    champ_dict['Role'].append(champs[x]['tags'])
    
champ_dict
```

```python
champ_df = pd.DataFrame(champ_dict)
champ_df.head(10)
```

```python
len(champ_df)
```

```python
roles = []

for x in range(len(champ_df)):
    ch_roles = champ_df.iloc[x, 2]
    
    for item in ch_roles:
        if item not in roles:
            roles.append(item)
    
    
roles
```

```python
champ_df["Fighter"] = 0
champ_df["Tank"] = 0
champ_df["Mage"] = 0
champ_df["Assassin"] = 0
champ_df["Support"] = 0
champ_df["Marksman"] = 0

champ_df.head()
```

```python
for x in range(len(champ_df)):
    ch_roles = champ_df.iloc[x, 2]
    
    for item in ch_roles:
        if item == "Fighter":
            champ_df.iloc[x, 3] = 1/(len(ch_roles))
        if item == "Tank":
            champ_df.iloc[x, 4] = 1/(len(ch_roles))
        if item == "Mage":
            champ_df.iloc[x, 5] = 1/(len(ch_roles))
        if item == "Assassin":
            champ_df.iloc[x, 6] = 1/(len(ch_roles))
        if item == "Support":
            champ_df.iloc[x, 7] = 1/(len(ch_roles))
        if item == "Marksman":
            champ_df.iloc[x, 8] = 1/(len(ch_roles))
            
champ_df.head(20)
        
```

```python
match_data = pd.read_csv("C:\\Users\\pmaxl\\OneDrive\\Desktop\\FinalGroup\\data_lol\\Dataframe2.csv")
match_data.rename(columns={'championId':'Key'}, inplace=True)
match_data['Key'] = match_data['Key'].astype(str)
len(match_data)
```

```python
merged_df = match_data.merge(champ_df, how = "left", on = "Key")
merged_df[0:10]
```

```python
match_df_list = []

n = int(len(merged_df)/10)

for x in range(n):
    
    count = x*10
    match = merged_df[count:(count+10)]
    match_df_list.append(match)
    


        

```

```python
len(match_df_list)

```

```python
match0 = match_df_list[0]
match0_w = match0[match0["win"] == True]
match0_w
```

```python
team_summaries = {
    "W/L": [],
    "Fighter": [],
    "Tank": [],
    "Mage": [],
    "Assassin":[],
    "Support": [],
    "Marksman": [],
    "vs Fighter": [],
    "vs Tank": [],
    "vs Mage": [],
    "vs Assassin": [],
    "vs Support": [],
    "vs Marksman": []
}

for x in range(len(match_df_list)):
    
    match = match_df_list[x]
    
    w = match[match['win'] == True]
    l = match[match['win'] == False]
    
    w_fighter = w['Fighter'].sum()
    w_tank = w['Tank'].sum()
    w_mage = w['Mage'].sum()
    w_assassin = w["Assassin"].sum()
    w_support = w["Support"].sum()
    w_marksman = w["Marksman"].sum()
    
    l_fighter = l['Fighter'].sum()
    l_tank = l['Tank'].sum()
    l_mage = l['Mage'].sum()
    l_assassin = l["Assassin"].sum()
    l_support = l["Support"].sum()
    l_marksman = l["Marksman"].sum()
    
    
    
    team_summaries['W/L'].append("w")
    team_summaries["Fighter"].append(w_fighter)
    team_summaries["Tank"].append(w_tank)
    team_summaries["Mage"].append(w_mage)
    team_summaries["Assassin"].append(w_assassin)
    team_summaries["Support"].append(w_support)
    team_summaries["Marksman"].append(w_marksman)
    team_summaries["vs Fighter"].append(l_fighter)
    team_summaries["vs Tank"].append(l_tank)
    team_summaries["vs Mage"].append(l_mage)
    team_summaries["vs Assassin"].append(l_assassin)
    team_summaries["vs Support"].append(l_support)
    team_summaries["vs Marksman"].append(l_marksman)
    
    
    team_summaries['W/L'].append("l")
    team_summaries["Fighter"].append(l_fighter)
    team_summaries["Tank"].append(l_tank)
    team_summaries["Mage"].append(l_mage)
    team_summaries["Assassin"].append(l_assassin)
    team_summaries["Support"].append(l_support)
    team_summaries["Marksman"].append(l_marksman)
    team_summaries["vs Fighter"].append(w_fighter)
    team_summaries["vs Tank"].append(w_tank)
    team_summaries["vs Mage"].append(w_mage)
    team_summaries["vs Assassin"].append(w_assassin)
    team_summaries["vs Support"].append(w_support)
    team_summaries["vs Marksman"].append(w_marksman)
    
    

summary_df = pd.DataFrame(team_summaries)
summary_df.head()
```
